```python
import pandas as  pd
```

```python
iPass=pd.read_csv("fact_TS_iPASS.csv")
iPass
```

```python
district=pd.read_csv("dim_districts.csv")
district
```

```python
iPassDistrict=pd.merge(iPass,district,on='dist_code',how='left')
iPassDistrict
```

```python
date=pd.read_csv("dim_date.csv")
date
```

```python
iPassDistrictDate=pd.merge(iPassDistrict,date,on='date',how='left')
iPassDistrictDate
```

```python
iPassDistrictDate.isnull().sum()
```

```python
iPassDistrictDate.to_csv('final_fact_iPass.csv', index=False)
```
