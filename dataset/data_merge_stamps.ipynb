```python
import pandas as  pd
```

```python
stamps=pd.read_csv("fact_stamps.csv")
stamps
```

```python
district=pd.read_csv("dim_districts.csv")
district
```

```python
stampDistrict=pd.merge(stamps,district,on='dist_code',how='left')
stampDistrict
```

```python
date=pd.read_csv("dim_date.csv")
date
```

```python
stampDistrictDate=pd.merge(stampDistrict,date,on='date',how='left')
stampDistrictDate
```

```python
stampDistrictDate.isnull().sum()
```

```python
stampDistrictDate.describe()
```

```python
stampDistrictDate.to_csv('final_fact_stamps.csv', index=False)
```
