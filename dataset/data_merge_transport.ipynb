```python
import pandas as  pd
```

```python
transport=pd.read_csv("fact_transport.csv")
transport
```

```python
district=pd.read_csv("dim_districts.csv")
district
```

```python
transportDistrict=pd.merge(transport,district,on='dist_code',how='left')
transportDistrict
```

```python
date=pd.read_csv("dim_date.csv")
date
```

```python
transportDistrictDate=pd.merge(transportDistrict,date,on='date',how='left')
transportDistrictDate
```

```python
transportDistrictDate.isnull().sum()
```

```python
transportDistrictDate.describe()
```

```python
transportDistrictDate.to_csv('final_fact_transport.csv', index=False)
```
